```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 600
a = 0.03
b = 0.0002
c = 0.06
d = 0.0001
```

```python
c/d*(a-b)
```

```python
def f(states):
    x, y = states
    x_next = (1+a)*x - b*x*y
    y_next = (1-c)*y + d*x*y
    states_next = np.array([x_next,y_next])
    return  states_next
```

```python
x_init = np.array([1600,100])

X = []
x_tmp = x_init
for t in range(N):
    x_tmp = f(x_tmp)
    X.append(x_tmp)

X = np.array(X)
plt.plot(X[:,0])
plt.plot(X[:,1])
```
