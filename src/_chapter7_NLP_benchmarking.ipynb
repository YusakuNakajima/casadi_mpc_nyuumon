```python
lbg = [-np.inf]
ubg = [0]
lbx = [-10,-6.5]
ubx = [0,0]

x0 = casadi.DM([-1,-1])

x = casadi.SX.sym("x")
y = casadi.SX.sym("y")

sinx = casadi.sin(x)
cosx = casadi.cos(x)
siny = casadi.sin(y)
cosy = casadi.cos(y)

f = siny*casadi.exp((1-cosx)**2) + cosx*casadi.exp((1-siny)**2) + (x-y)**2
g = (x+5)**2 + (y+5) - 25

nlp = {
    "x":casadi.vertcat(x,y),
    "f":f,
    "g":g
}
```

```python
S = casadi.nlpsol("S","ipopt",nlp,{"print_time":True})
res = S(x0=x0,lbg=lbg,ubg=ubg,lbx=lbx,ubx=ubx)
print(res)
```

```python
S = casadi.nlpsol("S_sqp","feasiblesqpmethod",nlp,{"print_time":True})
res = S(x0=x0,lbg=lbg,ubg=ubg,lbx=lbx,ubx=ubx)
print(res)
```
